A regressão linear mútipla trata de problemas em que existe uma relação linear entre múltiplas vatiáveis, de modo que a variável dependente é influenciada por todas essas variáveis.
Caso exista, essa função toma a forma de um polinômio de 1º grau como o somatório de fatores variáveis. Por exemplo: y = ax¹ + bx² + cx³ ... onde a, b, c são coeficientes e x¹, x², x³ ... são variáveis independentes.

Problemaa: qual investimento impacta mais no lucro de uma empresa: Administração, Pesquisa ou Propaganda? E o estado onde ela está inserida afeta em algo?

```python
#regressão linear múltipla
import pandas as pd
data  = pd.read_csv('startups.csv')
#definindo as variáveis
X = data.iloc[:, : -1].values
y = data.iloc[:, 4].values
data
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
import numpy as np
#Codificando a variável categórica estado
columnTransformer = ColumnTransformer([('encoder', OneHotEncoder(), [3])], remainder= 'passthrough')
X = np.array(columnTransformer.fit_transform(X), dtype = np.float)
X = X[:, 1:]
X
```

```python
#divisão da amostra com cross-validation
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

```python
#treinamento do algoritmo
from sklearn.linear_model import LinearRegression

#ajustando o modelo de Regressão Linear para o daraset de treino
regressor = LinearRegression()
regressor.fit(X_train, y_train)
```

```python
#prevendo o resultado do teste
y_pred = regressor.predict(X_test)
```

```python
for k in range(len(y_test)):
    y_teste = y_test[k]
    y_previsto = y_pred[k]
    err = abs((y_test[k] - y_pred[k])/ y_pred[k] *100)
    print(f'y_teste: {y_teste:.2f} y_previsto: {y_previsto:.2f} erro: {err:.2f}')
```

```python
#verificando os P-values de cada variável (quanto menor mais sgnificativo) 
#verificando a significância estatística das variáveis
import statsmodels.api as sm
X_opt = X[:,[2, 3]]
regressor_OLS = sm.OLS(y, X_opt).fit()
regressor_OLS.summary()
```

```python
regressor_OLS.pvalues
```
