Regressão polinomial é usada para representar problemas onde não há uma relação direta entre duas variáveis como na Regressão Linear. Aqui, a função é polinomial de grau superior a 1, uma parábola.
Há a possibilidade de a função ser um polinômio do segundo grau, por exemplo, y = -2x^4 + 2x³ + 3x² - 2x

```python
import pandas as pd
import matplotlib.pyplot as plt
data = pd.read_csv('cargo_nivel_salarios.csv')
data
```

```python
#separando o dataset em colunas X e y
X = data.iloc[:, 1:2].values
y = data.iloc[:, 2].values
print(X, y)
```

```python
#dividindo o dataset entre treino e teste
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

```python
#ajustando o modelo de Regressão linear para o dataset de treino
from sklearn.linear_model import LinearRegression
lin_reg = LinearRegression()
lin_reg.fit(X, y)
```

```python
#plotagem do gráfico com resultado do treino em regressão Linear
#Observa-se que o modelo errou pois os dados não formam uma reta no gráfico e sim uma curva
plt.plot(X, y, 'Dr')
plt.plot(X, lin_reg.predict(X), 'b-')
plt.title('Nivel X Salário (Regressão Linear)')
plt.xlabel('Nivel')
plt.ylabel('Salário')
plt.show()
```

```python
#criando o modelo de Regressão Polinomial
from sklearn.preprocessing import PolynomialFeatures
poly_reg = PolynomialFeatures(degree = 2)
```

```python
#ajustando o modelo para os dados de treino
X_poly = poly_reg.fit_transform(X_train)
poly_reg.fit(X_poly, y_train)
poly_reg.fit(X_train, y_train)
lin_reg_2 = LinearRegression()
lin_reg_2.fit(X_poly, y_train)
```

```python
#plotando os ados do treino
plt.figure(figsize = (15, 8))
plt.plot(X_train, y_train, 'Dr')
plt.plot(X_train, lin_reg_2.predict(poly_reg.fit_transform(X_train)), '^b')
plt.title('Nível x Salário (Regressão Polinomial de grau 2)')
plt.xlabel('Nível')
plt.ylabel('Salário')
plt.show()
```

```python
#aumentando o grau do polinômio para 4 ao invés de 2
poly_reg = PolynomialFeatures(degree = 4)
X_poly = poly_reg.fit_transform(X_train)
poly_reg.fit(X_poly, y_train)
poly_reg.fit(X_train, y_train)
lin_reg_3 = LinearRegression()
lin_reg_3.fit(X_poly, y_train)

#plotando os dados do teino
plt.figure(figsize=(15, 8))
plt.plot(X_train , y_train, 'Dr')
plt.plot(X_train, lin_reg_3.predict(poly_reg.fit_transform(X_train)), '^b')
plt.title('Nível x Salário (Regressão Polinomial de grau 4)')
plt.xlabel('Nível')
plt.ylabel('Salário')
plt.show()
```

```python
#plotando os dados do teste
plt.figure(figsize = (15, 8))
plt.plot(X_test, y_test, 'Dr')
plt.plot(X_test, lin_reg_3.predict(poly_reg.fit_transform(X_test)), '^b')
plt.title('Nível x Salário (Regressão Polinomial: TESTE)')
plt.xlabel('Nível')
plt.ylabel('Salário')
plt.show()

#amostras pobres visto que só temos duas para fazer os testes num 
#dataset com apenas 10 entradas
```

```python
#testando com todas as amostras
plt.figure(figsize = (15, 8))
plt.plot(X, y, 'Dr')
plt.plot(X, lin_reg_3.predict(poly_reg.fit_transform(X)), '^b--')
plt.title('Nível x Salário (Regressão Polinomial: TESTE GRAU 4)')
plt.xlabel('Nível')
plt.ylabel('Salário')
plt.show()
```

```python
#Modelo para teste com grau 8

#aumentando o grau do polinômio para 8 ao invés de 4
poly_reg = PolynomialFeatures(degree = 8)

#ajustando para todos os dados do dataset
X_poly = poly_reg.fit_transform(X)
poly_reg.fit(X_poly, y)
poly_reg.fit(X, y)
lin_reg_3 = LinearRegression()
lin_reg_3.fit(X_poly, y)

#testando com todas as amostras
plt.figure(figsize = (15, 8))
plt.plot(X, y, 'Dr')
plt.plot(X, lin_reg_3.predict(poly_reg.fit_transform(X)), '^b--')
plt.title('Nível x Salário (Regressão Polinomial: TESTE GRAU 8)')
plt.xlabel('Nível')
plt.ylabel('Salário')
plt.show()
```

```python
#Modelo para teste com grau 12
#Exemplo de modelo sobreajustado // overfitting

#aumentando o grau do polinômio para 12 ao invés de 8
poly_reg = PolynomialFeatures(degree = 12)

#ajustando para todos os dados do dataset
X_poly = poly_reg.fit_transform(X)
poly_reg.fit(X_poly, y)
poly_reg.fit(X, y)
lin_reg_3 = LinearRegression()
lin_reg_3.fit(X_poly, y)

#testando com todas as amostras
plt.figure(figsize = (15, 8))
plt.plot(X, y, 'Dr')
plt.plot(X, lin_reg_3.predict(poly_reg.fit_transform(X)), '^b--')
plt.title('Nível x Salário (Regressão Polinomial: TESTE GRAU 8)')
plt.xlabel('Nível')
plt.ylabel('Salário')
plt.show()
```
