```python
df = pd.read_csv('resources/week-1/datasets/Admission_Predict.csv', index_col = 0)
df.head()
```

```python
#agora vamos mudar o nome das colunas SPO e LOR pq ninguém sabe o que isso significa a não ser quem esta usando o DF
#no pandas usamos o rename() que aceita um parâmetro chamado colunas e precisamos passar um dicionário onde as chaves dele
#são os antigos nomes e os valores o novo nome da coluna
#old_name = new_name

new_df = df.rename(columns = {
    'GRE Score': 'GRE Score', 'TOEFL Score': 'TOEFL Score', 'University Rating': 'University Rating',
    'SOP': 'Statement of Pourpose', 'LOR ': 'Letter of Recomendation', 'CGPA': 'CGPA',
    'Research': 'Research', 'Chance of Admit ':'Chance of Admit'
})
new_df.head()
```

```python
new_df.columns
```

```python
#da pra usar um dicionário contendo somente o nome da coluna que queremos alterar
new_df = new_df.rename(columns = {'Letter of Recomendation': 'Letter Of Recomendation'})
new_df.head()
```

```python
#colocar um espaço no nome antigo funciona, mas e se fosse dois epsaços, um tab ou enter
#outra forma de arrumar o label das colunas é criar funções que fazem isso de forma automática
#renomeando as colunas de acordo. python vem com uma função mão na roda chamada srtip()
#quando usamos ela, nós passamos a função como parâmetro mapper e então indicamos se o eixo deve ser a coluna ou a linha
new_df = new_df.rename(mapper = str.strip, axis = 'columns')
new_df.head()
```

```python
#também podemos usar o df.columns atribudo-o á lista de nomes de colunas que renomeará as colunas
#isso vai mudar diretamente o df original, e é bom quando você tem várias colunas e quer usar só algumas

#como exemplo, vamos mudar o nome das colunas para letras minúsculas

cols = list(df.columns) #criando uma lista de labes das colunas
cols = [x.lower().strip() for x in cols] #transforme x em minúsculas e faça o strip para cada item x na lista cols
cols
```

```python
df.columns = cols #colunas do dataframe = lista com novos nomes
df.head()
```
