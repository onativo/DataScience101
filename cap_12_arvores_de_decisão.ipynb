Árvores de decisão (regressão e classificação), são considerados os melhores algoritmos de aprendizage de máquina.
Eles fornecem modelos com facilidade de interpretação e alta precisão.
Diferente do modelos lineares, eles conseguem modelar muito bem as relações não lineares, e sua implementação pode ser utulizada para vários problemas de classificação ou regressão.

**Vantagens:**
- Úteis na exploração dos dados, sendo uma das maneiras mais rápidas de identificar quais variáveis são significantes (têm relação com a variável dependente).
- Fáceis de endenter, pois sua representação gráfica é simples mesmo para aqueles sem conhecimento em estatística.
- Tratam tanto variáveis contínuas quanto categóricas, sem necessidade de limpeza, como outros modelos, sendo pouco afetadas poe valores anômalos e ausentes.
- Nâo necessitam de regra para criação da estrutura dos classificadores. Eles são criados a partir do próprio dataset.


**Desvantagens:**
- Apesar de poderem ser utilizadas com variáveis contínuas (numéricas), perdem informação contínua por discretizar variáveis em diferentes categorias
- Para evitar o sobreajuste, algumas restrições devem ser construídas e a árvore deve sofrer 'podas' em suas folhas

```python
#importando os dados da flor de iris
import matplotlib.pyplot as plt
from sklearn import datasets
iris = datasets.load_iris()
```

```python
#carregando o módulo Árvore de decisão da biblioteca sklearn
from sklearn import tree
```

O módulo 'tree' do sklearn recebe duas matrizes, a primeira, X é do tipo 'n * m' onde 'n' é a quantidade de amostras e 'm' é a quantidade de características de cada amostra.
A matriz Y é do tipo '1 * n', ou seja, uma matriz que tem uma única coluna contendo o valor inteiro indicanto a classe que cada amostra pertence.

No caso do dataset iris, a matriz X é do tipo 150*4, com 150 amostras de flores contendo 4 características núméricas: comprimento e largura da pétala e o comprimento e largura da sépala.
A segunda matriz, y, contém a mesma dimensão em linhas, 150, e com uma única coluna, indicando qual classe cada amostra percente, 0: setosa, 1: versicolor, 2: vergínica

```python
#carregando as matrizes X e y do dataset iris
X, y = iris.data, iris.target
```

```python
#criando um classificador do tipo árvore de decisão
classifier = tree.DecisionTreeClassifier()
#chamando o método fit()
classifier.fit(X, y)
```

```python
#setando a variável figsize do objeto plot
fig = plt.figure(figsize = (50, 45))
```

```python
#imprimindo a árvore
tree.plot_tree(classifier,
               feature_names = iris.feature_names,
               class_names = iris.target_names,
               filled = True)
```

```python
#gerando gráficos de dispersão
#agrupando por comprimento da sépala e comprimento da pétala
plt.figure(figsize = (15, 10))
plt.scatter(iris.data[:, 0], iris.data[:, 1], c = iris.target)
plt.xlabel(iris.feature_names[0])
plt.ylabel(iris.feature_names[2])
plt.tight_layout()
plt.show()
```

```python
#agrupando por comprimento da pétala e largura da pétala
plt.figure(figsize = (15, 10))
plt.scatter(iris.data[:, 0], iris.data[:, 1], c = iris.target)
plt.xlabel(iris.feature_names[2])
plt.ylabel(iris.feature_names[3])
plt.tight_layout()
plt.show()
```
